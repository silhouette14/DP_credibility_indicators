## Unpickle .pkl file back to dataframe

```python
#Raw data
data = pd.read_pickle("sources/df_articles.pkl")
#data = pd.read_pickle("sources/df_length_of_articles.pkl")

data.reset_index(inplace = True, drop = True)
data

```

```python
problem = 0
result = 0
for num in range(len(data)):
    try:
        url = data['url'][num]
        headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
        r = requests.get(url, headers=headers, verify=True, timeout=1)
        status = r.status_code
        print("{} Status code {}".format(num, status))
        print("URL {}".format(data['url'][num]))

        if r.status_code == 200:
            result = result + 1
        else:
            problem = problem + 1
    except:
        problem = problem + 1
        continue
    
```

```python
# REPORT
print("All sent requests: {}".format(problem+result))
print("Not responding/deleted articles {}".format(problem))
print("Working articles {}".format(result))
```
